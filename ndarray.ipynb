# 多维数组的切片和索引

ndarray 支持在多维数组上的切片操作。


创建一个数组并改变其维度,使之变成一个三维数组  

reshape 函数的作用是改变数组的“形状”,也就是改变数组的维度,其参数为一个正整数元组,分别指定数组在每个维度上的大小  
为了方便起见,我们可以用一个省略号( ... )来表示遍历剩下的维度。

```python
b = np.arange(96).reshape(8,3,4)
print(b)
```

## 切片操作

```python
print(b.shape)
print(b[0,:,:])
print(b[1, ...])
```

```python
print(b[3,1])
print(b[3,1,::2])   #在上面的数组切片中间隔地选定元素
print(b[...,2])
print(b[:,2])
print(b[0,:,2])
```

```python
print(b[:,:,-1])  #最后一列
```

```python
print(b[:,::-1,-1])  #反向选取最后一列
```

```python
print(b[::-1])    #翻转一维数组
print(b[...,::-1])
```

# 组合数组  
np.hstack() np.concatenate(axis=1)  水平组合  
np.vstack() np.concatenate(axis=0)  垂直组合  
np.dstack() 深度组合  
np.column_stack()   列组合  
np.row_stack()  行组合

```python
a = np.arange(9).reshape(3,3)
b = 2 * a
c = 2 * b
d = 2 * c
print(a)
print(b)
```

```python
np.hstack((a,b))   #水平组合
                   #将 ndarray 对象构成的元组作为参数,传给hstack 函数
```

```python
np.concatenate((a,b,c,d), axis=1)   #concatenate 函数来实现同样的效果
```

```python
np.vstack((a,b))    #垂直组合
```

```python
np.concatenate((a,b,c,d), axis=0)
```

```python
np.dstack((a,b))     #深度组合
                     #将一系列数组沿着纵轴(深度)方向进行层叠组合
                     #相同的元组作为参数传给 dstack 函数
```

```python
oned = np.arange(3)
twice_oned = 2 * oned
np.column_stack((oned, twice_oned)) #列组合
                                    #对于一维数组将按列方向进行组合
```

```python
np.column_stack((a,b)) == np.hstack((a,b)) #对于二维数组, column_stack 与 hstack 的效果是相同的
```

```python
np.row_stack((oned, twice_oned))    #行组合
                                    #一维数组,将直接层叠起来组合成一个二维数组
```

```python
np.row_stack((a,b)) == np.vstack((a,b))    #对于二维数组, row_stack 与 vstack 的效果是相同的
```

```python
g = b.flat  #flat 属性, 将返回一个 numpy.flatiter 对象
                                    #这是获得 flatiter 对象的唯一方式——我们无法访问 flatiter 的构造函数
print(g)
```

```python
for item in g:  #这个所谓的“扁平迭代器”可以让我们像遍历一维数组一样去遍历任意的多维数组
    print(item)
```

```python
print(b.flat[2])    #用 flatiter 对象直接获取一个数组元素
print(b.flat[[0,2]])    #或者获取多个元素
```

```python
b.flat[[0,2]] = 7   #flat 属性是一个可赋值的属性。对 flat 属性赋值将导致整个数组的元素都被覆盖
b
```
